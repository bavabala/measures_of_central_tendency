```python
import scipy as sp
train_a = [10,18,12,15,17]
train_b = [2,30,5,15,20]
```

```python
print('Mean delay of train_a:',sp.mean(train_a))
print('Mean delay of train_b:',sp.mean(train_b))
print('Median delay of train_a:',sp.median(train_a))
print('Median delay of train_b:',sp.median(train_b))
```

# Range


```python
print('Range of train_a:',max(train_a)-min(train_a))
print('Range of train_b:',max(train_b)-min(train_b)) # extreme values range is not good measure of variability
```

```python
sp.sqrt(2.66) # square root method
```

```python
print('std delay of train_a:',sp.std(train_a))
print('std delay of train_b:',sp.std(train_b)) #standard deviation
```

# Normality


```python
bp = [120,130,135,110,115,125,130,112,125]
```

```python
%matplotlib inline
import seaborn as sb
from scipy import stats
print("skewness of BP:",stats.skew(bp))
sb.distplot(bp)
```

```python
bp = [90,80,100,120,120,120,120,120,122,125,140,150,160,190]
print("skewness of BP:",stats.skew(bp))
print("kurtosis of BP:",stats.kurtosis(bp))
sb.distplot(bp)
```
